```python
meats=meat.head(120)
```

```python
q="""
Select
    strftime('%Y', date) as year,
    SUM(beef) as beef_total
    From
        meats
    Group by
        year
        Limit 5;
"""
pysqlfn(q)
```

```python
q="""
Select * from meats ORDER BY RANDOM() LIMIT 10
"""
pysqlfn(q)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

ds = pd.read_csv("https://stats.idre.ucla.edu/stat/data/binary.csv")
ds.hist()
```

```python
ds.plot(kind='density', subplots=True, layout=(2,2), sharex=False)
```

```python
ds.plot(kind='box', subplots=True, layout=(2,2), sharex=False, sharey=False)
```

```python
import numpy
correlations = ds.corr()
# plot correlation matrix
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = numpy.arange(0,4,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(ds.columns)
ax.set_yticklabels(ds.columns)
plt.show()
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(ds)
```
